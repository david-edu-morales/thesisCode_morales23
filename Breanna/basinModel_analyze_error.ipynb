**ADD ERROR TO OBSERVATIONS**

```python
if addError_flag == True:
    # Added measurement error to truth values in order to avoid concentrated likelihood on few models
    measurements = []       # define list to record truth values with obseravational error
    seed = 2                # seed pseudo-rand generator to simplify comparisons across stakeholders
    var  = 0.02             # set variance for obseravational error
    addedError = 1          # added error centered on mean = 0

    for i in range(len(truth_value)):
        np.random.seed(seed)
        measurement = abs(truth_value[i] * np.random.normal(1, var) + addedError * np.random.normal(0, var))
        measurements.append(measurement)
```
